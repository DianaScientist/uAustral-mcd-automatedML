```python
import pandas as pd
import numpy as np
import ast
```

```python
artists_genres = pd.read_csv('../input/spotifyartistsgenres/data_by_artist_o.csv')
artists_genres.head()
```

```python
genres_to_ignore = ['rock', 'pop', 'rap', 'dance pop', 'hip hop']
```

```python
genres = artists_genres['genres'].apply(lambda x: ast.literal_eval(x))
genres_ocurrences = pd.DataFrame(genres.explode().value_counts())
genres_ocurrences = genres_ocurrences.reset_index()
genres_ocurrences.columns = ['genre', 'count']
genres_ocurrences.head()
```

```python
def find_popular(genres_list):
    most_popular = ''
    most_popular_ocurrences = -1000
    for genre in genres_list:
        if (genres_ocurrences['genre'] == genre).any():
            count = int(genres_ocurrences[genres_ocurrences['genre'] == genre]['count'].iloc[0])
            if most_popular_ocurrences < count and genre not in genres_to_ignore:
                most_popular_ocurrences = count
                most_popular = genre
    return most_popular
```

```python
final_artists_genres = artists_genres.copy()
final_artists_genres = final_artists_genres[artists_genres['genres'].str.len() != 2]
final_artists_genres.head()
```

```python
final_artists_genres['popular_genre'] = final_artists_genres['genres'].apply(lambda x: find_popular(ast.literal_eval(x)))
```
